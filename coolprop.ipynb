```python
import numpy as np
from scipy.optimize import fsolve
def z(p):
  x,y = p
  return(x**2+x*y+7,y**2+np.sin(x)+10)
k = fsolve(z,(0,1)) 
print(k)
```

```python
import numpy as np
import CoolProp.CoolProp as ct
import matplotlib.pyplot as plt

t = np.linspace(274,380,100)
p = 101325
d = ct.PropsSI("D","T",t,"P",p,"water")
plt.plot(t,d,"r")
plt.scatter(373.15,ct.PropsSI("D","T",373.15,"P",p,"water"),color='black')
plt.title("график зависимости d от t при постоянном давление")
plt.xlabel("t")
plt.ylabel("d")
plt.grid()
plt.show()
```
